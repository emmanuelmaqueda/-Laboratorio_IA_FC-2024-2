## Función de Error

Cuando entrenas una red neuronal, generalmente tienes un conjunto de datos de entrada (por ejemplo, imágenes) y un conjunto de datos de salida esperados (etiquetas). La salida de la red neuronal es una predicción, y necesitas comparar esta predicción con la salida esperada para ajustar los pesos de la red y mejorar la precisión de las predicciones.

La función de error, también conocida como función de pérdida o función de costo, **cuantifica la discrepancia entre la salida predicha por la red y la salida real esperada. En otras palabras, mide qué tan "equivocada" está la red en sus predicciones.**

El paso desde la salida de la red neuronal a la función de error generalmente implica calcular la diferencia entre la salida predicha y la salida esperada para cada ejemplo de entrenamiento. Luego, esta diferencia se procesa a través de la función de error para obtener un valor único que representa cuán bien o mal está funcionando la red en general.

Existen diferentes funciones de error, cada una adecuada para diferentes tipos de problemas y situaciones. Por ejemplo, para problemas de clasificación, una función de error común es la entropía cruzada, mientras que para problemas de regresión, se utiliza a menudo el error cuadrático medio.

**El objetivo del entrenamiento de la red neuronal es minimizar esta función de error, ajustando los pesos de la red mediante técnicas de optimización como el descenso del gradiente. En resumen, la función de error es esencial en el proceso de entrenamiento de una red neuronal, ya que proporciona una medida cuantitativa de su rendimiento y guía la optimización de los pesos para mejorar las predicciones.**

En resumen:

 * El error se calcula entre las salidas esperadas y las salidas propagadas hacia adelante desde la red.

* Luego, estos errores se propagan hacia atrás a través de la red desde la capa de salida a la capa oculta, asignando culpas por el error y actualizando los pesos a medida que avanzan.

Backpropagation y cálculo del gradiente:

**El backpropagation es el proceso mediante el cual se calcula el gradiente de la función de costo con respecto a los pesos de la red neuronal.**

Durante la retropropagación, se propaga el error desde la capa de salida hasta la capa de entrada, calculando la contribución de cada peso en la red al error total.

Este cálculo se realiza utilizando la regla de la cadena y el gradiente de la función de activación en cada capa de la red.

**Descenso de gradiente y actualización de pesos:**

Una vez que se ha calculado el gradiente de la función de costo con respecto a los pesos, se utiliza el descenso de gradiente para ajustar los pesos de la red y minimizar la función de costo.

El descenso de gradiente implica actualizar los pesos en la dirección opuesta al gradiente multiplicado por una tasa de aprendizaje.

Esta actualización de pesos se realiza iterativamente para todos los pesos de la red, lo que gradualmente mejora el rendimiento de la red a medida que se minimiza la función de costo.

**En resumen, el backpropagation se encarga de calcular el gradiente de la función de costo, mientras que el descenso de gradiente se encarga de utilizar este gradiente para ajustar los pesos de la red. Juntos, estos dos procesos permiten que la red neuronal aprenda de los datos de entrenamiento y mejore su capacidad para realizar predicciones precisas.**
