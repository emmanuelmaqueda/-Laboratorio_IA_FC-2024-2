##El Invierno de la IA:


En 1956 se dio el término "inteligencia artificial" en Dartmouth durante una conferencia convocada por John McCarthy, a la cual asistieron, entre otros, Minsky, Newell y Simon. En esta conferencia se hicieron previsiones triunfalistas a diez años que jamás se cumplieron, lo que provocó el abandono casi total de las investigaciones durante quince años.

## Paradigma Simbólico o basado en conocimiento:

Características:
* Representación explícita del conocimiento utilizando símbolos y reglas.
* Uso de lenguajes de representación del conocimiento como lógica proposicional o lógica de primer orden.

Durante el invierno de la IA, los investigadores centraron su atención en el desarrollo de sistemas expertos. Los sistemas expertos utilizaban motores de inferencia basados en reglas para capturar el conocimiento humano y tomar decisiones en dominios específicos. Los sistemas expertos encontraron éxito en áreas como el diagnóstico médico y se convirtieron en una aplicación destacada de la IA durante este período.
